```python
correlation_mat = df.corr()
corr_pairs = correlation_mat.unstack()
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")
print('Negatively correlated pairs: ', sorted_pairs[sorted_pairs < 0])
print('Positively correlated pairs: ', sorted_pairs[sorted_pairs > 0.5])
```

```python
'''
Linear regression assumptions

1) linear relationship between variables - linearity test, scatter plot
2) independence of observations (Durbin-Watson test, value around 2)
3) homoscedasticity of the residuals (random distribution) - homogeneity of variance
4) no significant outliers (box whisker plot)
5) normal distribution of the residuals (normality test -> Shapiro Wilk >0.05)

'''

#checking for normality of independent variable (anonimyzed features) and dependent variable (target column)

Y = df['target']
X = df.drop('target', axis=1)

shapiro_X = stats.shapiro(X)
shapiro_Y = stats.shapiro(Y)

print('Shapiro-Wilk for independent variable: ', shapiro_X)
print('Shapiro-Wilk for dependent variable: ', shapiro_Y)
```

EDA conclusions:
* given dataset contains of 54 columns (float and integers) and 90 000 rows
* dataset does not have any unknown or missing values, almost each entry to the dataset is unique
* there are some negatively correlated pairs but only one positively (column 6 and column 8) - there is no need to drop any columns
* both independent and dependent variables are normally distributed (no need to transform data)

```python
# training

model = LinearRegression()

X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state = 40)

model = LinearRegression()
y_pred = model.fit(X_train,y_train).predict(X_test)

from sklearn.metrics import mean_squared_error

print('RMSE for training data: ', mean_squared_error(y_test, y_pred))
```

```python
print('R square for input and predicted value is: ', r2_score(y_test, y_pred)) # to check the direction of the relation (negative)
```

```python
# standarization of the data 
#to check if that will decrease the RMSE, which is high (meaning in nutshell that the expectations
# -> in nutshell are far away from reality, high value of the residuals)

from sklearn.linear_model import Lasso, Ridge

lasso = Lasso()
ridge = Ridge()

lasso.fit(X_train, y_train)
ridge.fit(X_train, y_train)

y_pred_lasso = lasso.predict(X_test)
y_pred_ridge = ridge.predict(X_test)

print('RMSE for Lasso: ', mean_squared_error(y_test, y_pred_lasso))
print('RMSE for Ridge: ', mean_squared_error(y_test, y_pred_ridge))
```
